```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

import tensorflow as tf
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
import joblib
```

```python
data = pd.read_csv('sales.csv')
```

Preprocessing

```python
def encode_dates(df, column):
    df[column] = pd.to_datetime(df[column], dayfirst=True)  # Use dayfirst=True to parse dates correctly
    df[f'{column}_year'] = df[column].dt.year
    df[f'{column}_month'] = df[column].dt.month
    df[f'{column}_day'] = df[column].dt.day
    df = df.drop(column, axis=1)
    return df

def onehot_encode(df, column):
    df = df.copy()
    dummies = pd.get_dummies(df[column], prefix=column)  # Corrected typo here
    df = pd.concat([df, dummies], axis=1)
    df = df.drop(column, axis=1)
    return df
```

```python
def preprocess_inputs(df):
  df = df.copy()

  #drop unnecessry columns
  df = df.drop(['Row ID', 'Customer Name', 'Country', 'Product Name'], axis=1)

  #drop  customer specific feature
  df = df.drop(['Order ID', 'Customer ID'], axis=1)

  #extract date features
  df = encode_dates(df, column='Order Date')
  df = encode_dates(df, column='Ship Date')

  #one-hot encode categorical features
  for column in ['Ship Mode', 'Segment', 'City', 'State', 'Postal Code', 'Region', 'Product ID', 'Category', 'Sub-Category']:
      df = onehot_encode(df, column=column)

  #split df into X an y
  y = df['Sales']
  X = df.drop('Sales', axis=1)

  #Train-test split
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.7, shuffle=True, random_state=1)

  #Scale X
  scaler = StandardScaler()
  scaler.fit(X_train)
  X_train = pd.DataFrame(scaler.transform(X_train), columns=X.columns)
  X_test = pd.DataFrame(scaler.transform(X_test), columns=X.columns)

 # Save the scaler for future use
  joblib.dump(scaler, 'scaler.joblib')

  return X_train, X_test, y_train, y_test

```

```python
X_train, X_test, y_train, y_test = preprocess_inputs(data)
```

**Training**

```python
inputs = tf.keras.Input(shape=(X_train.shape[1], ))
X = tf.keras.layers.Dense(1024, activation='relu')(inputs)
X = tf.keras.layers.Dense(1024, activation='relu')(X)
outputs = tf.keras.layers.Dense(1, activation='linear')(X)

model = tf.keras.Model(inputs=inputs, outputs=outputs)

print(model.summary())
```

```python
model.compile(
    optimizer='adam',
    loss='mse'
)

history = model.fit(
    X_train,
    y_train,
    validation_split=0.2,
    batch_size=32,
    epochs=100,
    callbacks=[
        tf.keras.callbacks.EarlyStopping(
                        monitor='val_loss',
                        patience=5,
                        restore_best_weights=True
                    ),
                    tf.keras.callbacks.ReduceLROnPlateau()
     ]
)
```

```python
# Save the trained model
model.save('sales_model.h5')
```

**Results**

```python
test_loss = model.evaluate(X_test, y_test, verbose=0)
print("Test Loss: {:.5f}".format(test_loss))
```

```python
y_pred = np.squeeze(model.predict(X_test))

test_r2 = r2_score(y_test, y_pred)

print("Test R^2 Score: {:.5f}".format(test_r2))
```

```python
# Function to load data from CSV
def load_data(csv_file):
    return pd.read_csv(csv_file)

# Function to make predictions with new data
def predict_sales(input_data, scaler, model):
    input_df = pd.DataFrame([input_data])
    input_df = pd.DataFrame(scaler.transform(input_df), columns=input_df.columns)
    prediction = model.predict(input_df)
    return prediction[0][0]

# Interface for user to enter data
def get_user_input():
    input_data = {}
    input_data['Order Date_year'] = int(input("Enter Order Year: "))
    input_data['Order Date_month'] = int(input("Enter Order Month: "))
    input_data['Order Date_day'] = int(input("Enter Order Day: "))
    input_data['Ship Date_year'] = int(input("Enter Ship Year: "))
    input_data['Ship Date_month'] = int(input("Enter Ship Month: "))
    input_data['Ship Date_day'] = int(input("Enter Ship Day: "))

    ship_mode = input("Enter Ship Mode: ")
    segment = input("Enter Segment: ")
    city = input("Enter City: ")
    state = input("Enter State: ")
    postal_code = input("Enter Postal Code: ")
    region = input("Enter Region: ")
    product_id = input("Enter Product ID: ")
    category = input("Enter Category: ")
    sub_category = input("Enter Sub-Category: ")

    input_data[f'Ship Mode_{ship_mode}'] = 1
    input_data[f'Segment_{segment}'] = 1
    input_data[f'City_{city}'] = 1
    input_data[f'State_{state}'] = 1
    input_data[f'Postal Code_{postal_code}'] = 1
    input_data[f'Region_{region}'] = 1
    input_data[f'Product ID_{product_id}'] = 1
    input_data[f'Category_{category}'] = 1
    input_data[f'Sub-Category_{sub_category}'] = 1

    for col in X_train.columns:
        if col not in input_data:
            input_data[col] = 0

    return input_data

# Load the trained model and scaler
model = load_model('sales_model.h5')
scaler = joblib.load('scaler.joblib')

# Get user input or load from CSV
choice = input("Do you want to load data from a CSV file? (yes/no): ").strip().lower()
if choice == 'yes':
    csv_file = input("Enter the path to the CSV file: ")
    user_data = load_data(csv_file)
    user_input = user_data.iloc[0].to_dict()  # Assuming the CSV has the correct structure
else:
    user_input = get_user_input()

# Make a prediction
predicted_sales = predict_sales(user_input, scaler, model)
print(f"Predicted Sales for the given data: {predicted_sales:.2f}")
```
